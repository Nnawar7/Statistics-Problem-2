```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simulate_central_limit(n):
    '''
    This function simulates the central limit theorem by simulating rolling a dice for n_rolls times and averaging the outcomes.
    Input:The number of dice which we throw.
    Output: histogram of the average outcomes, a vector with the average of each roll, 
    and the mean and variance of the total average
    outcomes.
    '''
    if n == 1:
        rolls =np.zeros (1000)
        for i in range (1000):
            rolls[i] = np.random.randint(1,7)
        plt.hist(rolls,bins=100)
        plt.grid()
        plt.title('Outcome of throwing a die 1000 times',fontsize=15)
        plt.xlabel('Outcomes',fontsize=20)
        plt.ylabel('Number of Occurence',fontsize=15)
        plt.show()
        mean = np.mean (rolls)
        var = np.var (rolls)
        avg_each_roll = mean
    elif n > 1:
        rolls = np.zeros (n * 1000)
        for i in range (n * 1000):
            rolls[i] = np.random.randint(1,7)
        sample = rolls.reshape (n,1000)  #each row indicates one sample
        y = np.zeros (1000)
        for index in range (n):
            y = y + sample [index,]
        avg_each_roll = y / n
        mean = np.mean(avg_each_roll)
        var = np.var(avg_each_roll)
        fig = plt.figure()
        axes1 = fig.add_axes([0.1, 0.1, 0.5, 0.5])
        axes1.hist(rolls,bins=100)
        axes1.set_xlabel('Outcomes')
        axes1.set_ylabel('Number of Occurence')
        axes1.set_title('Outcome of throwing {} dice 1000 times'.format(n))
        axes2 = fig.add_axes([0.7, 0.5, 0.6, 0.5])
        axes2.hist(avg_each_roll,bins=100)
        axes2.set_xlabel('Average of outcomes')
        axes2.set_ylabel('Number of Occurance')
        axes2.set_title('average of the resulting values of {} dice in each time.'.format(n))
        plt.show()
    
    return avg_each_roll, mean, var
```

```python
avgs1, mean1, var1 = simulate_central_limit(1)
avgs2, mean2, var2 = simulate_central_limit(2)
avgs3, mean3, var3 = simulate_central_limit(10)

print("In case of rolling the die once: ")
print("Mean = {0:f}, Variance = {1:f}".format(mean1, var1))
print("Average of outcomes in each case: ", avgs1, "\n")

print("In case of rolling the TWO dice: ")
print("Mean = {0:f}, Variance = {1:f}".format(mean2, var2))
print("Average of outcomes in each case: ", avgs2, "\n")

print("In case of rolling the TEN dice: ")
print("Mean = {0:f}, Variance = {1:f}".format(mean3, var3))
print("Average of outcomes in each case: ", avgs3, "\n")
```
